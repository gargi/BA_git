# Section kNN

```python
def kNN(train_data,train_labels,test_data,timed=False,n_jobs = 4,k = 20,thresh = 0.8):
    #performance-testing
    if timed:
        start = time.time()
    
    #setup
    neigh = neighbors.KNeighborsClassifier()
    neigh.set_params(algorithm = 'auto',
                     leaf_size = 30,
                     metric = 'minkowski',
                     metric_params = None,
                     n_jobs = n_jobs,
                     n_neighbors = k,
                     p = 2,
                     weights = 'distance')
    
    neigh.fit(train_data,train_labels)
    
    #predict test_data, use threshold
    soft_pred = neigh.predict_proba(test_data)
    hard_pred = tb.customThreshold(soft_pred[:,1],thresh)
    
    if timed:
        end = time.time()
        print(end - start, "seconds")
    
    #
    del neigh
    return hard_pred,soft_pred

# Times for 
# 20.928184986114502 sec - k=20
# 264.2239580154419 sec - k=5
```

```python
hard_pred_1 = kNN(train_1,train_labels,test_1,timed=True,k=10)[0]
b_ams,v_ams = tb.calcSetAMS(hard_pred_1,sol_data)
print("Public AMS:",b_ams[0],"|| Private AMS:",v_ams[0])
```

```python
#tb.createSubmissionFile(test_events,hard_pred_1,"test.csv",threshold = 0.5)
```

```python
hard_pred_all = kNN(train_all,train_labels,test_all,timed=True,k=100)[0]
b_ams,v_ams = tb.calcSetAMS(hard_pred_all,sol_data)
print("Public AMS:",b_ams[0],"|| Private AMS:",v_ams[0])
```

```python
soft_pred_2 = kNN(train_2,train_labels,test_2,timed=True,k=20)[1]

hard_pred_2,maxAMS,maxThresh = tb.bestThreshold(soft_pred_2[:,1],sol_data)

b_ams,v_ams = tb.calcSetAMS(hard_pred_2,sol_data)
print("Public AMS:",b_ams[0],"|| Private AMS:",v_ams[0],"|| Threshold:",maxThresh)
```

# Section logistic Regression

```python
def logisticReg(train_data,train_labels,test_data,timed=False,n_jobs = 4,thresh = 0.42):
    #performance-testing
    if timed:
        start = time.time()
    
    #setup
    logReg = linMod.LogisticRegression()
    logReg.set_params(C= 1.0,
                  class_weight= None,
                  dual= False,
                  fit_intercept= True,
                  intercept_scaling= 1,
                  max_iter= 100,
                  multi_class= 'ovr',
                  n_jobs= n_jobs,
                  penalty= 'l2',
                  random_state= None,
                  solver= 'liblinear',
                  tol= 0.0001,
                  verbose= 0,
                  warm_start= False)
    
    #training classifier
    logReg.fit(train_data,train_labels)
    logReg.sparsify()
    
    #predict test_data, use threshold
    soft_pred = logReg.predict_proba(test_data)
    hard_pred = tb.customThreshold(soft_pred[:,1],thresh)
    
    if timed:
        end = time.time()
        print(end - start, "seconds")
    
    #
    del logReg
    return hard_pred,soft_pred
```

```python
hard_pred_all = logisticReg(train_all,train_labels,test_all,timed=True,thresh = 0.42)[0]
b_ams,v_ams = tb.calcSetAMS(hard_pred_all,sol_data)
print("Public AMS:",b_ams[0],"|| Private AMS:",v_ams[0])
```

```python
hard_pred_all = logisticReg(train_1,train_labels,test_1,timed=True,thresh = 0.42)[0]
b_ams,v_ams = tb.calcSetAMS(hard_pred_all,sol_data)
print("Public AMS:",b_ams[0],"|| Private AMS:",v_ams[0])
```
